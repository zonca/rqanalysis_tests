## Horizontal join

In horizontal join instead we have 2 dataframes that share the same index, at least part of it, and we want to align the columns.
Here I am sampling 80% of the rows and then sorting them, so that we do not have identical indices.

```python
## If index is the same, join is crazy fast

%time _ = data1.join(data1.copy(), rsuffix="copy")
```

```python
data1_80perc = data1.sample(frac=.8).sort_index()
```

```python
%memit data1_join_data1 = data1.join(data1_80perc, rsuffix="copy")
```

```python
data1_join_data1.info()
```

```python
%time data1_join_data1 = data1.join(data1_80perc, rsuffix="copy")
```

```python
data1_join_data1
```

## Horizontal join with no multiindex

Check if multi-index is making joins slower, it doesn't seem so

```python
data1_nomultiindex = data1.reset_index(drop=True) 
data1_80perc_nomultiindex = data1_80perc.reset_index(drop=True) # I drop the index so we should have the same amount of memory
```

```python
%time data1_join_data1 = data1.join(data1_80perc, rsuffix="copy")
```

```python
%memit data1_join_data1_no_multiindex = data1_nomultiindex.join(data1_80perc_nomultiindex, rsuffix="copy")
```

The memory profiler result I think are not very accurate, not sure why this operation should take 530 MB while the same with multi index 400MB, less than the size of the new dataframe!
